```python
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
data_encoded= pd.get_dummies(df)
clf = IsolationForest(random_state =10, contamination=0.1)
clf.fit(data_encoded)
#Prediction
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

```python
# Add an 'Outlier' column to the dataframe for easy identification
df['Outlier'] = y_pred_outliers

# Step 5: Remove the outliers (keep only inliers)
df_inliers = df[df['Outlier'] == 1]
df_inliers
```
